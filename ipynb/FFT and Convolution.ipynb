```python
import numpy as np
import scipy as sp
a = np.linspace(1,4)
y = lambda x: x**2
z = lambda x: np.log(x)
```

```python
from scipy.fftpack import fft, ifft
from scipy.signal import convolve
nmap = lambda f,l: np.array(list(map(f,l)))
y_ = nmap(y,a)
z_ = nmap(z,a)
yz_ = nmap(lambda t: y(t)*z(t), a)
```

```python
import pylab as plt
plt.plot(a, yz_) # fg
plt.plot(a, convolve(fft(y_), fft(z_), 'same'))
plt.show()
```

```python
plt.plot(a, ifft(fft(yz_))) # fg
plt.plot(a, ifft(fft(y_)*fft(z_))) # f*g = Finv(F(f)F(g))
plt.plot(a, ifft(fft(fft(y_))*fft(fft(z_)))) # F(f)*F(g) = Finv(F(F((f))F(F(g)))
plt.plot(a, ifft(ifft(fft(fft(y_))*fft(fft(z_))))) # fg = Finv(F(f)*F(g)) = Finv(Finv(F(F(f))F(F(g))))
plt.show()
```
